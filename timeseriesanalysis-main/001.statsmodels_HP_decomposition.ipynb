## Hodrick Prescott Filter

### The filter is used for decomposition of the time series in two components
#### 1. The trend part
#### 2. The cyclical part

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Data/macrodata.csv', index_col='Unnamed: 0', parse_dates=True)
```

```python
df.head()
```

```python
# plotting the two trend alongwith the original data
df['realgdp'].plot(figsize=(12,6))
```

```python
# Now this Hodrick Prescott filter is used to seperate the time series component into trend part and the cyclical part

from statsmodels.tsa.filters.hp_filter import hpfilter
df['cycle'], df['trend'] = hpfilter(df['realgdp'], lamb=1600)
```

```python
df['cycle'], df['trend'] = hpfilter(df['realgdp'], lamb=1600)
```

```python
# plotting the two trend alongwith the original data
df[['realgdp','trend']].plot(figsize=(12,6));
```

```python
#plotting the cyclical component of the data
df['cycle'].plot(figsize=(12,6));
```

### We have seperated the given time series into two components and plotted
