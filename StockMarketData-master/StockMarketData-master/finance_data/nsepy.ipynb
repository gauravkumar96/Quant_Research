### In this notebook we'll explore the usage of the python library nsepy 
    1. Getting the historical data for stocks
    2. Getting the bhawcopy data
    3. Getting historical data for futures
    4. Getting historical data for options
    5. Other Miscellaneous features

```python
# getting the stock historical data
from nsepy import get_history
from datetime import date
data = get_history(symbol="SBIN", start=date(2015,1,1), end=date(2015,1,31))
data.head()
```

```python
# Stock futures (Similarly for index futures, set index = True)
sbin_fut = get_history(symbol="SBIN",
                        start=date(2015,1,1),
                        end=date(2015,1,10),
                        futures=True,
                        expiry_date=date(2015,1,29))
sbin_fut.head()
```

```python
# Stock options (Similarly for index options, set index = True)
stock_opt = get_history(symbol="SBIN",
                        start=date(2015,1,1),
                        end=date(2015,1,10),
                        option_type="CE",
                        strike_price=300,
                        expiry_date=date(2015,1,29))
stock_opt.head()
```

```python
# Index Historical data
nifty_50 = get_history(symbol="NIFTY 50",
                            start=date(2015,1,1),
                            end=date(2015,1,10),
                            index=True)
nifty_50.head()
```

```python
# Index futures data
nifty_fut = get_history(symbol="NIFTY",
                        start=date(2015,1,1),
                        end=date(2015,1,10),
                        index=True,
                        futures=True,
                        expiry_date=date(2015,1,29))
nifty_fut.head()
```

```python
# Index Options data
nifty_opt = get_history(symbol="NIFTY",
                        start=date(2015,1,1),
                        end=date(2015,1,10),
                        index=True,
                        option_type='CE',
                        strike_price=8200,
                        expiry_date=date(2015,1,29))
nifty_opt.head()
```

```python
# VIX historical data
vix = get_history(symbol="INDIAVIX",
            start=date(2015,1,1),
            end=date(2015,1,10),
            index=True)
vix.head()
```

```python
# fetching the expiry dates
from nsepy.derivatives import get_expiry_date
expiry = get_expiry_date(year=2020,month=4)
print(expiry)
```

```python
# Index P/E ratio history
from nsepy import get_index_pe_history
nifty_pe = get_index_pe_history(symbol="NIFTY",
                                start=date(2015,1,1),
                                end=date(2015,1,10))
nifty_pe
```

```python
# Getting the RBI reference 
from nsepy import get_rbi_ref_history
rbi_ref = get_rbi_ref_history(date(2015,1,1), date(2015,1,10))
rbi_ref
```

```python
# Getting the bhaw copy of the data
from nsepy.history import get_price_list
prices = get_price_list(dt=date(2015,1,1))
prices.head()
```
